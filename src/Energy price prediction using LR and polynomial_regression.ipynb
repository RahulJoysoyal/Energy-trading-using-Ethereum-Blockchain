## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Demand_Supply_Data.csv')
dataset.head()
```

```python
X = dataset.iloc[:, 0:3].values
y = dataset.iloc[:, -1].values
```

## Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
plt.scatter(X[:, 0], y, color='red')  # Plotting the first column of X against y
plt.plot(X[:, 0], lin_reg.predict(X), color='blue')  # Plotting the predicted values using linear regression
plt.title('Price according to energy supply,demand,timestamp using Linear Regression')
plt.xlabel('Energy supply,Demand,Timestamp')
plt.ylabel('Price')
plt.show()
```

## Training the Polynomial Regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 10)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

## Visualising the Linear Regression results

```python
plt.scatter(X[:, 0], y, color='red')  # Plotting the first column of X against y
plt.plot(X[:, 0], lin_reg.predict(X), color='blue')  # Plotting the predicted values using linear regression
plt.title('Price according to energy supply,demand,timestamp using Linear Regression')
plt.xlabel('Energy supply,demand,timestamp')
plt.ylabel('Price')
plt.show()
```

## Visualising the Polynomial Regression results

```python
plt.scatter(X[:,0], y, color = 'red')
plt.plot(X[:,0], lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Price according to energy supply,demand,timestamp using Polynomial Regression')
plt.xlabel('Energy supply,demand,timestamp')
plt.ylabel('Price')
plt.show()
```

## Predicting a new result with Linear Regression

```python
x = lin_reg.predict([[1689404960,6354,6530]])
x
```

## Predicting a new result with Polynomial Regression

```python
y = lin_reg_2.predict(poly_reg.fit_transform([[1689404960,6354,6530]]))
y
```

## Optimal Value

```python
minimum_value = np.minimum(x, y)
print("Optimum price is:",minimum_value)
```
